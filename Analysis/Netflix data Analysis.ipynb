```python
# assigns new columns for weekday and hour
title_copy = title.copy()
title_copy['weekday'] = title_copy['Start Time'].dt.weekday
title_copy['hour'] = title_copy['Start Time'].dt.hour
title
```

```python
# set the Categorial and tell the order to be plotted from Monday to Sunday
title_copy['weekday'] = pd.Categorical(title_copy['weekday'], categories=[0, 1, 2, 3, 4, 5, 6], ordered=True)
title_copy['weekday'] = title_copy['weekday'].cat.rename_categories(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# counts the rows and assign it to title_by_day
title_by_day = title_copy['weekday'].value_counts()

# Sorts the index using our categorical, so that Monday (0) is first, Tuesday (1) is second, etc.
title_by_day = title_by_day.sort_index()

# update the font to a viewable size
plt.rcParams.update({'font.size': 26})

# defining the x and y axisf
plt.xlabel('DAYS OF THE WEEK')
plt.ylabel('NUMBER OF EPISODES')

# Plot title_by_day as a bar chart with the listed size and title
title_by_day.plot(kind='bar', figsize=(25, 12), title='BAR CHART OF STAR TREK EPISODES WATCHED BY DAY')

# executes the graph
plt.show()
```

```python
# This code shows a line graph to track changes in number of episodes watched based on days
plt.rcParams.update({'font.size': 26})

plt.figure(figsize=(20,10))

plt.plot(title_by_day.index, title_by_day.values, marker='o', color = 'red')

plt.xlabel('DAYS OF THE WEEK')
plt.ylabel('NUMBER OF EPISODES')
plt.title('LINE GRAPH OF STAR TREK EPISODES WATCHED BY DAY')

plt.show()
```

```python
# displays the data of episodes watched by day in form of pie chart
matplotlib.rcParams.update({'font.size': 19})
title_by_day.plot(kind='pie', figsize=(20, 10), title='Pie Chart of Star Trek Episodes Watched by Day', autopct='%1.1f%%')
plt.show()
```

```python
# This displays continuous data by days of week with respect to change in epidode

# updates the font to a viewable size
plt.rcParams.update({'font.size': 26})

plt.figure(figsize=(20,10))

# This depicts the bars of the histogram plot
plt.hist(title_copy['weekday'], bins=len(title_by_day), color = 'orange')

# Labels both axis and title
plt.xlabel('DAYS OF THE WEEK')
plt.ylabel('FREQUENCY')
plt.title('HISTOGRAM OF STAR TREK EPISODES WATCHED BY DAY')

plt.show()
```

```python
# This code depicts a bar chart that compares distribution of episodes watched based on the hour of the day
title_copy['hour'] = pd.Categorical(title_copy['hour'], categories=[i for i in range(24)], ordered=True)

# counts the rows and assign it to title_by_day
title_by_hour = title_copy['hour'].value_counts()

# Sort the index using our categorical, so that midnight (0) is first, 1 a.m. (1) is second, etc.
title_by_hour = title_by_hour.sort_index()

plt.rcParams.update({'font.size': 25})

plt.xlabel('HOURS OF DAY')
plt.ylabel('NUMBER OF EPISODES')

title_by_hour.plot(kind='bar', figsize=(35, 15), title='BAR CHART OF STAR TREK EPISODES WATCHED BY HOUR', color = 'purple')
plt.show()
```
